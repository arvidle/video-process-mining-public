```python
# Fix frame filenames
import os
rootpath = "/content/ava_dataset/frames/"
folders = [
"836-0107-0108",
"836-0108-0109",
"836-0109-0110",
"836-0113-0114",
"836-0114-0115",
"836-0125-0126",
"931-0015-0016",
"931-0035-0036",
"931-0036-0037",
"946-0021-0022",
"946-0022-0023",
"946-0023-0024",
"946-0024-0025",
"954-0000-0001"
]
for folder in folders:
  folderpath = rootpath + folder + "/"
  for filename in os.listdir(folderpath):
    video, suffix = filename.split("_")
    frame_id, ext = suffix.split(".")
    new_filename = "img_" + str(int(frame_id)).zfill(5) + "." + ext
    os.rename(folderpath + filename, folderpath + new_filename)
```
